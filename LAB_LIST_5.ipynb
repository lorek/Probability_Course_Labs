# Rozgrzewka: słońce i deszcz

Załóżmy, że dni bezchmurne (z bezchmurnym niebem przez co najmniej 16 godzin) stanowią 45% wszystkich dni w ciągu roku, natomiast dni deszczowe (te, w których w ogóle wystąpił opad) 40% wszystkich dni. Czy można stąd wywnioskować, że 18% spośród dni w ciągu roku to dni słoneczne i deszczowe zarazem? Oczywiście nie, bo zachmurzenie i opady nie są od siebie niezależne. Każda kombinacja tych dwóch zjawisk ma swoje prawdopodobieństwo, które najprościej przedstawić w tabeli:

| zachmurzenie | bez opadów | z opadami |
|--------------|--------------|--------------|
| brak               | 0.41              | 0.04              |
| częściowe      | 0.18              | 0.12              |
| pełne              | 0.09              | 0.16              |

Tworzymy macierz prawdopodobieństw dla stanów pogody. Łączny rozkład prawdopodobieństwa przedstawimy na wykresie słupkowym.

```python
import numpy as np
import matplotlib.pyplot as plt


zachmurzenie = ["słonecznie", "trochę chmur", "pochmurnie"]
opady = ["bez opadów", "z opadami"]
prawdopodobienstwa = np.array([[.41, .04], [.18, .12], [.09, .16]])

kolory = ["gold", "skyblue", "slategray"]


x = np.arange(len(opady))
width = 0.25  # szerokosc slupka
offsets = np.linspace(-width, width, len(zachmurzenie))


for idx, (cloud, color, offset) in enumerate(zip(zachmurzenie, kolory, offsets)):
    #print("idx = ", idx, " x + offset, prawdopodobienstwa[idx], width=width -> ",x + offset, prawdopodobienstwa[idx], width)
    plt.bar(x + offset, prawdopodobienstwa[idx], width=width, label=cloud, color=color)


plt.xticks(x, opady)
plt.xlabel("Opady")
plt.ylabel("Prawdopodobieństwa")
plt.title("Prawdopodobieństwa w zależności od zachmurzenia i opadów")


plt.legend(loc="upper right")


plt.show()
```
