````python
# more customization

x = np.arange(-2,2, 0.2)

y1= 4*np.sin(x)
y2 = np.exp(x)


plt.plot(x, y1, label='4*sin(x)') # each label automatically different color
plt.plot(x, y2, label='exp(x)')
plt.legend()

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('4*sin(x) and exp(x)')

plt.show()
````
